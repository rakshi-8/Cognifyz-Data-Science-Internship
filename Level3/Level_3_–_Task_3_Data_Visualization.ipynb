```python
import pandas as pd

df = pd.read_csv("dataset.csv")

df = df.dropna(subset=['Aggregate rating', 'Cuisines', 'City', 'Votes', 'Price range'])
df['Aggregate rating'] = df['Aggregate rating'].astype(float)
df['Votes'] = df['Votes'].astype(int)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['Aggregate rating'], bins=20, kde=True, color='green')
plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
top_cuisines = df['Cuisines'].value_counts().head(10).index

top_df = df[df['Cuisines'].isin(top_cuisines)]

plt.figure(figsize=(10,6))
sns.barplot(x='Cuisines', y='Aggregate rating', data=top_df, estimator='mean', palette='coolwarm')
plt.title("Average Aggregate Rating by Top 10 Cuisines")
plt.xticks(rotation=45)
plt.show()
```

```python
top_cities = df['City'].value_counts().head(10).index
city_df = df[df['City'].isin(top_cities)]

plt.figure(figsize=(10,6))
sns.barplot(x='City', y='Aggregate rating', data=city_df, estimator='mean', palette='magma')
plt.title("Average Rating by Top 10 Cities")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Votes', y='Aggregate rating', data=df, color='orange')
plt.title("Votes vs Aggregate Rating")
plt.xlabel("Votes")
plt.ylabel("Aggregate Rating")
plt.show()
```
